```python
print("======STUDENT PROFILE SYSTEM======")
name = input("Enter Full Name: ")
age_input  = input("Enter Age: ")
gpa_input  = input("Enter GPA: ")
courses_input = input("Enter courses (separated by commas): ")


age = int(age_input)
gpa = float(gpa_input)
courses = courses_input.split(",") 
num_courses = len(courses)         


student_profile = {
    "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": num_courses
}


is_deans_list = gpa >= 4.0
is_adult = age >= 18


deans_list_text = "Yes" if is_deans_list else "No"
age_text = "Adult" if is_adult else "Minor"


print("\n--- Type Conversions ---")
print(f"Age: {type(age_input)} became {type(student_profile['age'])}")
print(f"GPA: {type(gpa_input)} became {type(student_profile['gpa'])}")
print(f"Courses: {type(courses_input)} became {type(student_profile['courses'])}")


print("\n===== STUDENT PROFILE =====")
print(f"Name:        {student_profile['name']}")
print(f"Age:         {student_profile['age']} ({age_text})")
print(f"GPA:         {student_profile['gpa']}")
print(f"Courses:     {student_profile['courses']}")
print(f"# Courses:   {student_profile['num_courses']}")
print(f"Dean's List: {deans_list_text}")
print("===========================")
```
